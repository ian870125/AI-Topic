

### VC BOUND

$\epsilon(h) < \hat{\epsilon}(h) + \sqrt{\frac{8}{n}log(\frac{4(2n)^d}{delta})}$<br>
n: number of training instance, d: VC Dimension(model complexity)

### VC Dimension

* Model Complexity
* Categorized the hypothesis set into finite groups


* Growth Function
$$ \tau(x_1, \dots, x_n) = \vert \{(h(x_1), \dots, h(x_n))\vert h \in H\}\vert$$
$$ \tau_H(n) = max_{(x_1, \dots, x_n)}\tau (x_1, \dots, x_n)$$

* VC Dimension 
$$ d(H) = max\{n : \tau_H(n) = 2^n\}$$



* VC Dimension is independent from data distribution

* VC Dimension of linear model:
    * O(W)
    * W = number of parameters 
* VC Dimension of fully-connected neural networks:
    * O(LW log )
    * L = number of layers
    * W = number of parameters
    




### PAC-Bayesian Bound
PAC-Bayesian Bound is data-dependent
* Learning is feasible when $\hat{\epsilon}(Q)$ is small $\rightarrow \epsilon (Q)$ is small
* With 1 - $\delta$ probability, the following inequality (PAC-Bayesian Bound) is satisfied
$$ \epsilon(Q) \leq \hat{\epsilon}(Q) + \sqrt{\frac{KL(Q\Vert P) + \log (\frac{n}{\delta}) + 2}{2n - 1}}$$



PAC-Bayesian Bound for Deep Learning
(Stochastic)<br>
With 1-$\delta$ probability, the following inequality is satisfied
$$ KL(\hat{\epsilon}(Q)\Vert \epsilon(Q)) \leq \frac{KL(Q\Vert P) + \log(\frac{n}{\delta})}{n - 1}$$
Q: SNN after training, P: SNN before training

Constrain on sharpness: given a margin $\Upsilon > 0$, if Q satisfy:
$$P_{h' \backsim Q} \{ sup_{x \in X}\Vert h'(x) - h(x)\Vert_\infty \leq \frac{\Upsilon}{4}\} \geq \frac{1}{2}$$


With 1 - $\delta$ probability, the following inequality is satisfied.
$$ L_0(h) \leq \hat{L}_\Upsilon(h) + 4\sqrt{\frac{KL(Q\Vert P) + \log(\frac{6n}{\delta})}{n - 1}}$$
margin loss $L_{\Upsilon}(h)$:
$$L_{\Upsilon}(h) = P_{(x, y)\backsim D}\{h(x)[y] \leq \Upsilon + max_{j \neq y}h(x)[j]\}$$
